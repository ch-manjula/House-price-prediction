Modify the provided Python code to perform linear regression on the '/content/diamonds.csv' dataset using the 'carat' and 'price' columns.

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load dataset (or upload your own)
df = pd.read_csv('/content/diamonds.csv')  # You can also use a housing CSV

# Choose numeric features
df = df[['carat', 'price']]
df.dropna(inplace=True)

# Train/test split
X = df[['carat']]
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model training
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))

# Plot
plt.scatter(X_test, y_test, label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.legend()
plt.show()
```
